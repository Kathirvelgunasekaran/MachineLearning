```python
#import pandas, it's Globalized function, under pandas what are all the function exit it'll be loaded into file created memory
#Pandas we are using for database related Table creation and editing. Full table supporting
#under pandas 1000 functions there means it'll loaded into memory. more consumption will be there
#shift+enter using to run programms
#once we hae executed below statement In[] this cell change like In[1] and next execution time It'll change like In[2]
#import pandas as pd - henceforth we are going to use pandas as pd. pd it'll reflect as pandas
#pandas is a libray it contains functions related to the database
import pandas as pd
```

```python
#Now we are goingto load dataset
#Always make sure dataset "Salary_Data.csv" file exist under respective folder.otherwise it'll through File not found error.
#Try to avoid spelling mistakes.
dataset=pd.read_csv("Salary_Data.csv") #dataset is called by pd.read_csv function under pandas and assigned to the dataset varaiable.
dataset #now we are seeing what is there under dataset
#dataset is called by pd.read_csv function under pandas and assigned to the dataset varaiable
```

```python
#Now going to perform Input and output variables split
#In the above table YearsExperience is input and Salary is output
#Here Data preprocessing will not cover, already we have preprocessed data, In this model we are going to learn how to use preprocessd data like that
#inside dataset varaiable we have YearsExperience and Salary values.
#we are going to assign YearsExperience column into varaiable Independent. same like Salary column value assign into Dependent varaiable.
independent=dataset[["YearsExperience"]]
independent #Now YearsExperience value seperately assigned into independent varaiable, Independent means input, Dependent means outptu
```

```python
dependent=dataset[["Salary"]] #Now Salary value seperately assigned into dependent varaiable,
dependent
#Now we got seperatley independent and dependent varaiables. now splited input and output
```

```python
#Now split Trainset and Testset. 
#Using sklearn library going to split Trainset and Testset. it's supporting for ML.
#respective folder(skleran) under this function(model_selection) we are going to (import) or calling train_test_split
#Not calling all functions under respective folder(skleran) under this function(model_selection) we are going to call train_test_split.
#under pandas 1000 functions there means it'll loaded into memory. more consumption will be there
#here just calling sklearn folder under model_selection we are calling train_test_split function only. due to this kind of calling memory consumption will be low
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) #No varaiable assigned here.here paraleely split input and output based on 70:30 ratio.
#It's parameterized function, now we are going to pass Independent(input) and Dependent(output) varaiable. test_size=0.30(we are dividing this test_size ratio rate 70:30, usually Trainset should be high compartively Testset).
#Here we are getting four varaiale output like list.it throwing four varaiable results.this function train_test_split based on four varaiable return.
#we have to create four varaiable like(X_train,X_test,y_train,y_test)
#we are using _ instead of space into vraiables(X_train,X_test,y_train,y_test)
#X means input, y means output, X input based trainset and testset(X_train,X_test) and y based output trainset and testset(y_train,y_test) 
#you can use sklearn library based on trainset and testset split like this four variables only(X_train,X_test,y_train,y_test).
#As of now compltetd Trainset and Testset split.
```

```python
#Now we are going to see Linear regression algorithm libary functions and input passing that we are going to see now
#ML algorithm all are provided by sklearn 
from sklearn.linear_model import LinearRegression # While calling, this line loaded into memory. here LinearRegression may be function or class. In this line LinearRegression considered as class
regressor=LinearRegression() #Non parameterized function we are assigning into regressor varaiable.it consider as class, all the function under this class assign to regressor.
regressor.fit(X_train,y_train) #fit function we are calling it using access operator, now passing splited trainset input and output.
#here just calling sklearn folder under linear_model we are calling LinearRegression function only. due to this kind of calling memory consumption will be low
#without spelling mistake do it
#Top companies Regression creating means they set model varaiable like Regressor nd following for the same.
#first from line loaded into memory, now second line will execute
#what are all the functions LinearRegression under this class loaded into regressor varaable.
#class and functions are just we can call it as procedure.
#Third line when we call fit method and passing materials(X_train,y_train). this formula respective place substitute values.
#First two lines are procedure.we have holded weight and bias values, while calling fit() after execution this line it'll calculate weight and bias.now Model is ready
#For this particular block have Learning phase going to happen here,
#Secret is, All ML algorithm just these lines path only we are going to change,,remaining all codes of all algorithms are same.
#---------------------------------------------------------------
#from sklearn.linear_model import LinearRegression # 
#regressor=LinearRegression()
#---------------------------------------------------------------
#regressor.fit(X_train,y_train) - while this line execution completed means model creation was completed. respective constant values based on dataset calculaed .
```

```python
#Now we have to see Module calculated value.
weight=regressor.coef_  #coef_ method using to calculate weight. while calling this "regressor.coef_" it will save it into weight varaiable 
weight #checking weight saved value.weight/slope is next prediction hook. we should know how pattern going.
```

```python
bais=regressor.intercept_  #intercept_ method using to calculate bais. while calling this "regressor.intercept" it will save it into bais varaiable 
bais #checking bais saved value.
#regressor.fit(X_train,y_train) -In this regressor all are saved(formula,weight,bais)
#Now we know weight and bais
```

```python
#Evaluation metrix(R2square using)
y_pred=regressor.predict(X_test) #y-pred is varaiable, all are saved in regressor(formula,weight,bais), predict is application function.
#Evaluation matrix shold happened on Testdataset .Testset values anyone not included into Traunset values. so that we are using it for crosss predition
#y-test is Actual data(y_test), X_test is brain predicted outptut.so that we are using it for crosss predition
```

```python
from sklearn.metrics import r2_score #using this r2_score function going to find r2 value.
r_score=r2_score(y_test,y_pred) #r2_score assigned to r_score variable, now passing Actual output(y_test) and predicted output(y_pred)
#regressor=LinearRegression()
#r2_score nearly 1 means Good model, nearly to 0 means bad model
#Now it nearly to 1. our model is performing well. our model learned well while performing crosscheck. evaluation matrix cross checking our model and model is performing well or not. 
```
